```python
import numpy as np
import pandas as pd
```

```python
#import dataset
df = pd.read_csv('Data_SPP.csv')
```

```python
df.info()
```

```python
for column in df.columns:
    print(column)
    print(*sorted(df[column].unique()))
```

```python
#we will consider ssc_p between 50 to 100
df = df[(df['ssc_p'] >= 50 ) & (df['ssc_p']<= 100)]
```

```python
df.info()
```

```python
df['ssc_p'].unique()
```

```python
#we will consider hsc_p between 40 to 100
df = df[(df['hsc_p'] >= 40) & (df['hsc_p'] <= 100)]
```

```python
df.info()
```

```python
#handle salary column
df['salary'] = df['salary'].replace(0,np.nan)
median = df['salary'].median()
df['salary'] = df['salary'].fillna(median)
```

```python
df.info()
```

```python
df = df[df['degree_t'] != 'Others']
```

```python
df.info()
```

```python
df.to_csv('cleaned_data.csv', index = False)
```
