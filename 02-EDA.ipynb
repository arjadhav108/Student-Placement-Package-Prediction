```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
#import cleaned dataset
df = pd.read_csv('cleaned_data.csv')
```

```python
#analysis
df.info()
```

```python
#gender distribution
plt.figure(figsize = (6,4))
sns.countplot(x = 'gender', data = df)
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("No of students")
plt.show()
```

```python
#hsc_s wise hsc_p
sns.swarmplot(x = 'hsc_s', y = 'hsc_p', data = df)
plt.xticks(rotation = 90)
plt.show()
```

```python
#hsc_s wise degree_p
ax = df['hsc_s'].value_counts().sort_index().plot(kind = 'bar')
ax.bar_label(ax.containers[0],fontsize = 6)
plt.title("HSC_S wise Degree Percentage")
plt.xlabel('hsc_s')
plt.ylabel('degree_p')
plt.show()
```

```python
#workex wise count of students
df['workex'].value_counts().plot(kind = 'pie', autopct = "%1.1f", figsize = (5,5), startangle = 90)
plt.title("Work-Experience")
plt.show()
```

```python
#degree_t wise salary
sns.swarmplot(x = 'degree_t', y = 'salary', data = df)
plt.xticks(rotation = 90)
plt.show()
```

```python
#degree_p wise salary
plt.figure(figsize = (8,5))
sns.scatterplot(x = 'degree_p', y = 'salary', data = df)
plt.title("Degree Percentage wise Salary")
plt.xlabel("Degree Percentage")
plt.ylabel("Salary")

plt.show()
```

```python
#hsc_p distribution
plt.figure(figsize = (8,5))
sns.histplot(df['ssc_p'], bins = 10)
plt.title("SSC Percentage Distribution")
plt.xlabel("SSC Percentage")
plt.ylabel("No of students")
plt.show()
```
